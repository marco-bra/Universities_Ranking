# 4. For each ranking, consider only the most recent data point. For each university, compute the maximum difference between the rankings (e.g. for *Aarhus University* the value is 122-73=49). Notice that some rankings are expressed as a range

Alcune posizioni vengono espresse sottoforma di intervallo (es. 101-150) nelle classfiche di *shanghai* e *Times*; quest'ultima presenta inoltre il carattere "=" per indicare un pari merito. Per questo è stata creata una funzione che permette di indivuare la parte numerica e riportare la media tra gi estremi nel caso in cui sia presente un intervallo di posizioni, in modo da avere un valore indicativo su cui poter calcolare la differenza massima tra i 3 ranking.

```python
def range_to_mean (dt, column):
    dt[['inf', 'sup']] = dt[column].str.extract('(\d+)-?(\d+)?')
    dt['inf'] = pd.to_numeric(dt['inf'])
    dt['sup'] = dt['sup'].fillna(dt['inf'])
    dt['sup'] = pd.to_numeric(dt['sup'])
    dt['rank'] = round( (dt['sup'] + dt['inf']) /2)
    del dt['inf']
    del dt['sup']
    return dt
```

Viene applicata la funzione ai dati più recenti di *Times* e viene verificato che la colonna risultante sia di tipo decimale.

```python
recent_times = range_to_mean(recent_times, 'world_rank')
print(recent_times['rank'].head())
recent_times.head()
```

Vengono estratti i dati più recenti di *shanghai*, viene applicata la funzione e verificato che la colonna risultante sia di tipo decimale.

```python
recent_shanghai = shanghai.iloc[shanghai.groupby('univ_name')['year'].idxmax()].copy()
recent_shanghai = range_to_mean(recent_shanghai, 'world_rank')
print(recent_shanghai['rank'].head())
recent_shanghai.head()
```

## Unione dei ranking

La tabella risultante deve avere tutte le università, di conseguenza l'unione delle tabelle deve restituire un numero di righe $\geq$ alla tabella con più righe.

```python
print(recent_times.shape)
print(recent_shanghai.shape)
print(recent_cwur.shape)
```

Si uniscono le tabelle dei 3 ranking sulla chiave `key` utilizzando la modalità *outer*, cioè inserendo nella tabella finale sia le università in comune, sia quelle non in comune. <br>
Si verifica inoltre che il numero di righe è 1291 $\geq$ 1024.

```python
fuso_temp = pd.merge(recent_times, recent_shanghai, on = 'key', how = 'outer', suffixes = ('_times', '_shanghai'))
fuso = pd.merge(fuso_temp, recent_cwur, on = 'key', how = 'outer', suffixes = ('', '_cwur'))
fuso = fuso.rename(columns = {'world_rank':'rank_cwur'})
print(fuso.shape)
fuso[['university_name_times', 'rank_times', 'rank_shanghai', 'rank_cwur']].head()
```

Una volta ottenute le posizioni in classifiche dei dati più recenti per ogni università, è stata creata una funzione in grado di calcolare la differenza massima tra questi, ovviando al problema dei valori mancanti creati creati dall'unione delle tabelle. <br>
Nel caso in cui 2 dei 3 ranking siano valori mancanti, viene restituito *NaN*

```python
def diff_rank (a, b, c):
    diff = float()
    if (pd.notna(a) & pd.notna(b) & pd.notna(c)):
        diff = max(a, b, c) - min(a, b, c)
    elif (pd.notna(a) & pd.notna(b)):
            diff = max(a, b) - min(a, b)
    elif (pd.notna(a) & pd.notna(c)):
            diff = max(a, c) - min(a, c)
    elif (pd.notna(b) & pd.notna(c)):
            diff = max(b, c) - min(b, c)
    else:
        diff = float('NaN')
    return diff
```

Viene applicata la funzione creata precedentemente riga per riga, in modo da creare una nuova colonna corrispondente alla differenza più grande tra le 3 posizioni.

```python
fuso['diff_max'] = fuso.apply(lambda row: diff_rank(row['rank_times'], row['rank_shanghai'], row['rank_cwur']), axis = 1)
fuso.sort_values(by = 'diff_max', inplace = True, ascending = False)
fuso[['university_name_times', 'country', 'rank_times', 'rank_shanghai', 'rank_cwur', 'diff_max']].dropna()
```

> L'università *Zhejiang University* dell'Arabia Saudita prensenta la differenza massima più elevata tra i ranking *shanghai* e *Cwur*, mentre 2 università americane e una inglese presentano differenze massime di una sola posizione tra le classifiche.

# 13. The universities that, in any year, have the same position in all three rankings (they must have the same position in a year).

Ad ogni università delle classifiche di partenza viene associato il valore corrispondente della chiave creata al punto 4 tramite una *left join*.

```python
times['key'] = times['univ_name']

shanghai = pd.merge(shanghai, s_t_change[['input_list_item', 'similar_ref_list_item']], left_on = 'univ_name', right_on = 'input_list_item',  how = 'left')
shanghai = shanghai.rename(columns = {'similar_ref_list_item':'key'})
shanghai['key'] = shanghai['key'].fillna(shanghai['univ_name'])

cwur = pd.merge(cwur, c_ts_change[['input_list_item', 'similar_ref_list_item']], left_on = 'univ_name', right_on = 'input_list_item',  how = 'left')
cwur = cwur.rename(columns = {'similar_ref_list_item':'key'})
cwur['key'] = cwur['key'].fillna(cwur['univ_name'])
```

Per poter individuare le università che hanno la stessa posizione in classifica nello stesso anno, si è optato per una unione di tipo *inner*. Come riportato in vari punti, la classifiche *Times* e *shanghai* riportano alcune posizioni in classifica come intervallo: con il metodo precedente si dovrebbero avere intervalli definiti per tutte le classifiche, cosa che in realtà non avviene. Per tenere in considerazione anche i range, si è deciso di riportare anche quelle università che potrebbero avere stessa posizione in tutte e 3 le classifiche.

Viene effettuata una nuova unione dei ranking con chiave la coppia `key`-`year`, in modo da avere per ognuna i 3 posizionamenti nei diversi ranking.

```python
fusion_temp = pd.merge(cwur, times, on = ['key', 'year'], suffixes = ('', '_t'))
range_fusion = pd.merge(fusion_temp, shanghai, on = ['key', 'year'], suffixes = ('', '_s'))
range_fusion.rename(columns = {'world_rank':'WR_Cwur', 'world_rank_t':'WR_Times', 'world_rank_s':'WR_shanghai'}, inplace = True)
range_fusion[['university_name', 'year', 'WR_Cwur', 'WR_Times', 'WR_shanghai']]
```

Si crea una funzione che permette di estrarre l'estremo inferiore e l'estremo superiore degli intervalli. Nel caso di valori puntuali viene riportato lo stesso numero sia come estremo inferiore sia come estremo superiore.

```python
def range_to_inf_and_sup (dt, column, inf, sup):
    dt[[inf, sup]] = dt[column].str.extract('(\d+)-?(\d+)?')
    dt[inf] = pd.to_numeric(dt[inf])
    dt[sup] = dt[sup].fillna(dt[inf])
    dt[sup] = pd.to_numeric(dt[sup])
    return dt
```

Applicando la funzione alle 2 colonne che presentano valori intervallari, vengono create le colonne contenenti gli estremi di ognuna.

```python
range_fusion = range_to_inf_and_sup(range_fusion, 'WR_Times', 'WRt_inf', 'WRt_sup')
range_fusion = range_to_inf_and_sup(range_fusion, 'WR_shanghai', 'WRs_inf', 'WRs_sup')
range_fusion['WR_Cwur'] = pd.to_numeric(range_fusion['WR_Cwur'])
range_fusion[['university_name', 'year', 'WR_Cwur', 'WR_Times', 'WRt_inf', 'WRt_sup', 'WR_shanghai', 'WRs_inf', 'WRs_sup']]
```

A questo punto, avendo le posizioni della classifica *Cwur* esclusivamente in maniera puntuale, si possono considerare quelle università il cui valore di `WR_Cwur` sia contemporaneamente maggiore degli estremi inferiori e minore di quelli superiori di `WR_Times` e `WR_shanghai`.

```python
range_fusion[(range_fusion['WR_Cwur'] >= range_fusion['WRt_inf']) 
             & (range_fusion['WR_Cwur'] >= range_fusion['WRs_inf'])
             & (range_fusion['WR_Cwur'] <= range_fusion['WRt_sup'])
             & (range_fusion['WR_Cwur'] <= range_fusion['WRs_sup'])][['university_name', 'country', 'year', 'WR_Cwur', 'WR_Times', 'WR_shanghai']]
```
