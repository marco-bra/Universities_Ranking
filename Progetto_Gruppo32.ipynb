# 4. For each ranking, consider only the most recent data point. For each university, compute the maximum difference between the rankings (e.g. for *Aarhus University* the value is 122-73=49). Notice that some rankings are expressed as a range

Alcune posizioni vengono espresse sottoforma di intervallo (es. 101-150) nelle classfiche di *shanghai* e *Times*; quest'ultima presenta inoltre il carattere "=" per indicare un pari merito. Per questo √® stata creata una funzione che permette di indivuare la parte numerica e riportare la media tra gi estremi nel caso in cui sia presente un intervallo di posizioni, in modo da avere un valore indicativo su cui poter calcolare la differenza massima tra i 3 ranking.

```python
def range_to_mean (dt, column):
    dt[['inf', 'sup']] = dt[column].str.extract('(\d+)-?(\d+)?')
    dt['inf'] = pd.to_numeric(dt['inf'])
    dt['sup'] = dt['sup'].fillna(dt['inf'])
    dt['sup'] = pd.to_numeric(dt['sup'])
    dt['rank'] = round( (dt['sup'] + dt['inf']) /2)
    del dt['inf']
    del dt['sup']
    return dt
```

Viene applicata la funzione ai dati pi√π recenti di *Times* e viene verificato che la colonna risultante sia di tipo decimale.

```python
recent_times = range_to_mean(recent_times, 'world_rank')
print(recent_times['rank'].head())
recent_times.head()
```

Vengono estratti i dati pi√π recenti di *shanghai*, viene applicata la funzione e verificato che la colonna risultante sia di tipo decimale.

```python
recent_shanghai = shanghai.iloc[shanghai.groupby('univ_name')['year'].idxmax()].copy()
recent_shanghai = range_to_mean(recent_shanghai, 'world_rank')
print(recent_shanghai['rank'].head())
recent_shanghai.head()
```

## Unione dei ranking

La tabella risultante deve avere tutte le universit√†, di conseguenza l'unione delle tabelle deve restituire un numero di righe $\geq$ alla tabella con pi√π righe.

```python
print(recent_times.shape)
print(recent_shanghai.shape)
print(recent_cwur.shape)
```

Si uniscono le tabelle dei 3 ranking sulla chiave `key` utilizzando la modalit√† *outer*, cio√® inserendo nella tabella finale sia le universit√† in comune, sia quelle non in comune. <br>
Si verifica inoltre che il numero di righe √® 1291 $\geq$ 1024.

```python
fuso_temp = pd.merge(recent_times, recent_shanghai, on = 'key', how = 'outer', suffixes = ('_times', '_shanghai'))
fuso = pd.merge(fuso_temp, recent_cwur, on = 'key', how = 'outer', suffixes = ('', '_cwur'))
fuso = fuso.rename(columns = {'world_rank':'rank_cwur'})
print(fuso.shape)
fuso[['university_name_times', 'rank_times', 'rank_shanghai', 'rank_cwur']].head()
```

Una volta ottenute le posizioni in classifiche dei dati pi√π recenti per ogni universit√†, √® stata creata una funzione in grado di calcolare la differenza massima tra questi, ovviando al problema dei valori mancanti creati creati dall'unione delle tabelle. <br>
Nel caso in cui 2 dei 3 ranking siano valori mancanti, viene restituito *NaN*

```python
def diff_rank (a, b, c):
    diff = float()
    if (pd.notna(a) & pd.notna(b) & pd.notna(c)):
        diff = max(a, b, c) - min(a, b, c)
    elif (pd.notna(a) & pd.notna(b)):
            diff = max(a, b) - min(a, b)
    elif (pd.notna(a) & pd.notna(c)):
            diff = max(a, c) - min(a, c)
    elif (pd.notna(b) & pd.notna(c)):
            diff = max(b, c) - min(b, c)
    else:
        diff = float('NaN')
    return diff
```

Viene applicata la funzione creata precedentemente riga per riga, in modo da creare una nuova colonna corrispondente alla differenza pi√π grande tra le 3 posizioni.

```python
fuso['diff_max'] = fuso.apply(lambda row: diff_rank(row['rank_times'], row['rank_shanghai'], row['rank_cwur']), axis = 1)
fuso.sort_values(by = 'diff_max', inplace = True, ascending = False)
fuso[['university_name_times', 'country', 'rank_times', 'rank_shanghai', 'rank_cwur', 'diff_max']].dropna()
```

> L'universit√† *Zhejiang University* dell'Arabia Saudita prensenta la differenza massima pi√π elevata tra i ranking *shanghai* e *Cwur*, mentre 2 universit√† americane e una inglese presentano differenze massime di una sola posizione tra le classifiche.

# 7. For each country, compute the fraction of the students in the country that are in one of the universities computed in the previous point (that is, the denominator of the ratio is the total number of students over all universities in the country).

Per calcolare il numeratore dell'assegnamento, vengono considerate solo le universit√† che hanno un rapporto $\frac{\textit{Femmine}}{\textit{Maschi}}$ minore della media, individuate nella tabella ottenuta al punto precedente dai dati pi√π recenti di *Times*. <br>
Viene calcolata la somma di tutti gli studenti raggruppati per `country`.

```python
stud_ratio = times_ratio.groupby('country', as_index = False)['num_students'].sum()
stud_ratio.rename(columns = {'num_students':'country_students_ratio'}, inplace = True)
stud_ratio.head()
```

Per calcolare il numeratore dell'assegnamento, vengono considerate tutte le universit√† riferite ai dati pi√π recenti di *Times*. <br>
Viene calcolata la somma di tutti gli studenti raggruppati per `country`. <br>
Le somme pari a 0 vengono considerate come *NaN*, in modo da non scaturire un errore causato dal denominatore di una frazione nullo. Queste molto probabilmente corrispondono ad errori di inserimento dati o comunque a valori mancanti.

```python
all_stud_for_country = recent_times.groupby('country', as_index = False)['num_students'].sum()
all_stud_for_country.rename(columns = {'num_students':'all_country_students'}, inplace = True)
all_stud_for_country.replace(0, float('NaN'), inplace = True)
all_stud_for_country
```

Viene eseguita un'unione con chiave `country` di tipo "right" tra la prima tabella ottenuta e la seconda ripulita dai valori mancanti, per assicurarsi che vengano riportate tutte le nazioni con almeno uno studente di *Times*. Dopoidch√® viene sostituito il valore 0 laddove venga generato un *NaN*, in modo che venga comunque calcolata una percentuale, seppur pari a 0.

```python
students_country = pd.merge(stud_ratio, all_stud_for_country.dropna(), on = 'country', how = 'right')
students_country['country_students_ratio'] = students_country['country_students_ratio'].fillna(0)
students_country['perc'] = round(students_country['country_students_ratio'] / students_country['all_country_students'] * 100, 2)
students_country[['country', 'country_students_ratio', 'all_country_students', 'perc']].sort_values(by = 'perc', ascending = False)
```

# 13. The universities that, in any year, have the same position in all three rankings (they must have the same position in a year).

Come riportato in vari punti, la classifiche *Times* e *shanghai* riportano alcune posizioni in classifica come intervallo: con il metodo precedente si dovrebbero avere intervalli definiti per tutte le classifiche, cosa che in realt√† non avviene. Per tenere in considerazione anche i range, si √® deciso di riportare anche quelle universit√† che potrebbero avere stessa posizione in tutte e 3 le classifiche.

Viene effettuata una nuova unione dei ranking con chiave la coppia `key`-`year`, in modo da avere per ognuna i 3 posizionamenti nei diversi ranking.

```python
fusion_temp = pd.merge(cwur, times, on = ['key', 'year'], suffixes = ('', '_t'))
range_fusion = pd.merge(fusion_temp, shanghai, on = ['key', 'year'], suffixes = ('', '_s'))
range_fusion.rename(columns = {'world_rank':'WR_Cwur', 'world_rank_t':'WR_Times', 'world_rank_s':'WR_shanghai'}, inplace = True)
range_fusion[['university_name', 'year', 'WR_Cwur', 'WR_Times', 'WR_shanghai']]
```

Si crea una funzione che permette di estrarre l'estremo inferiore e l'estremo superiore degli intervalli. Nel caso di valori puntuali viene riportato lo stesso numero sia come estremo inferiore sia come estremo superiore.

```python
def range_to_inf_and_sup (dt, column, inf, sup):
    dt[[inf, sup]] = dt[column].str.extract('(\d+)-?(\d+)?')
    dt[inf] = pd.to_numeric(dt[inf])
    dt[sup] = dt[sup].fillna(dt[inf])
    dt[sup] = pd.to_numeric(dt[sup])
    return dt
```

Applicando la funzione alle 2 colonne che presentano valori intervallari, vengono create le colonne contenenti gli estremi di ognuna.

```python
range_fusion = range_to_inf_and_sup(range_fusion, 'WR_Times', 'WRt_inf', 'WRt_sup')
range_fusion = range_to_inf_and_sup(range_fusion, 'WR_shanghai', 'WRs_inf', 'WRs_sup')
range_fusion['WR_Cwur'] = pd.to_numeric(range_fusion['WR_Cwur'])
range_fusion[['university_name', 'year', 'WR_Cwur', 'WR_Times', 'WRt_inf', 'WRt_sup', 'WR_shanghai', 'WRs_inf', 'WRs_sup']]
```

A questo punto, avendo le posizioni della classifica *Cwur* esclusivamente in maniera puntuale, si possono considerare quelle universit√† il cui valore di `WR_Cwur` sia contemporaneamente maggiore degli estremi inferiori e minore di quelli superiori di `WR_Times` e `WR_shanghai`.

```python
range_fusion[(range_fusion['WR_Cwur'] >= range_fusion['WRt_inf']) 
             & (range_fusion['WR_Cwur'] >= range_fusion['WRs_inf'])
             & (range_fusion['WR_Cwur'] <= range_fusion['WRt_sup'])
             & (range_fusion['WR_Cwur'] <= range_fusion['WRs_sup'])][['university_name', 'country', 'year', 'WR_Cwur', 'WR_Times', 'WR_shanghai']]
```
